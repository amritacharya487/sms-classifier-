```python
from nltk.corpus import stopwords
```

```python
import string 
```

```python
def text_tokanize(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    
    
    y =[]
    for i in text:
        if i.isalnum():
          y.append(i)
    return y
```
